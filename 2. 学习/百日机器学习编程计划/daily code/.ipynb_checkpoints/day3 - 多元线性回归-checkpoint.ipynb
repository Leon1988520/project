# 多元线性回归

## 基本假设
1. 自变量和因变量存在且只存在线性关系
2. 方差齐性(homoscedasticity)
3. 误差项正态分布(multivariate normality)
4. 不存在完全共线性(lack of multicollinearity)


[多元线性回归假设](https://blog.csdn.net/omandrake/article/details/49391427)

[如何理解回归中的方差齐性](https://www.sohu.com/a/197715883_655370)

## 基本过程
1. data preprocessiong
2. fit train set
3. predict test set


## keynote

> 如果数据为字符型的，在做OnehotEncoder之前，要先进行LabelEncoder，OnehotEncoder无法对字符型变量进行热编码

## 1. data preprocessing

```python
# import library
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# import dataset
data = pd.read_table("../data/day3-50startups.txt"
                     , sep = ","
                    )

X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values
data.head()
```

```python
# encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lblencoder = LabelEncoder()

X[:, 3] = lblencoder.fit_transform(X[:, 3])
X
```

```python
onehotencoder = OneHotEncoder(categorical_features=[3])
X = onehotencoder.fit_transform(X).toarray()
X
```

```python
X
```

```python
# aviod dummy variable trap

X = X[:, 1:]
```

```python
# split the dataset into training and test data
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

## 2. 模型拟合

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression().fit(X_train, Y_train)
```

# 3. 模型预测

```python
y_pred = regressor.predict(X_test)
```

# 4. 模型评估

```python
(Y_test - y_pred) / Y_test
```
