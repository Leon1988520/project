    ```

    ```

2. 下载Neo4j zip文件（个人使用，请下载community版本）；将其解压到目标文件夹下；在环境变量中配置Neo4j_HOME

3. 启动Neo4j程序

    - 以管理员身份打开控制台(WIN+R, 键入cmd)，切到Neo4j的bin文件下，运行： neo4j.bat console
    > 出现 neo4j.bat console'Powershell' 不是内部或外部命令，也不是可运行的程序或批处理文件

    - 解决方案：在环境变量path后添加上powershell的路径（C:\Windows\System32\WindowsPowerShell\v1.0），重新打开cmd窗口按照教程运行就可以了
    
    以上完成了Neo4j的启动，可以在第4步，将其注册为服务

4. 把Neo4j安装为服务
    ```
    安装和卸载服务：
    bin\neo4j install-service
    bin\neo4j uninstall-service

    bin\neo4j start
    bin\neo4j stop
    bin\neo4j restart
    bin\neo4j status
    ```

## 五. 使用Neo4j

> 使用Cyber进行数据建模, Cypher是描述性的图形查询语言，语法简单，功能强大，由于Neo4j在图形数据库家族中处于绝对领先的地位，拥有众多的用户基数，使得Cypher成为图形查询语言的事实上的标准。本文在Neo4j Browser中执行Cypher示例代码。本节内容：

### 5.1. 基本概念

  - variable
    > 指示类型，是实体、关系还是其他，一般实体使用n表示，关系使用r
    > 表示变量，建立关系时，作为中间变量
    
  - Label
    > 表示类，如实体Person、Movie、关系Relationship
    
    > 所以 variable:Label刻画了被描述的类型以及对象，如n:Person，描述了实体类型的人这一种对象
  - key
    > 在Cypher数据模型中，没有实例的概念，key、value对组成的就是实例的属性
  - value
  
### 5.2 基本操作

1. 创建节点
  > 节点模式的构成 variable:Label1[:Label2]{Key1:Value1, Key2:Value2}

  ```
    create (n:Director{name:"张艺谋", gender:"男性"}) return n;
    create (n:Actor{name:"梁朝伟", gender:" 男性"}) return n;
    create (n:Actor{name:"张曼玉", gender:"女性"}) return n;
    create (n:Movie{title:"英雄", dt_publish:2001}) return n;
  ```

  > 每个节点都有一个整数ID，在创建新节点时，Neo4j会自动为设点设置ID值，在整个数据库中，节点的ID是递增且唯一的

2. 查询节点

  - 查询整个图数据库 match(n) return n;
  - 查询特定属性值(key)的节点 match(n) where n.gender = "男性" return n;
  - 查询具有指定属性值的节点 match(n{name:"张艺谋"}) return n; # 和上面的语句是等价的
  - 查询具有指定Label的节点 match(n:Movie) return;
  
3. 创建关系

  > 关系的构成 startnode - [variable:relationshipType{Key1:Value1, Key2:value2}] -> endnode
  > 创建关系时，要先将关系两端的实体(上面说Cypher中不存在实体，是不存在表示实体的字段，实体本身是存在的)查询出来，再在此基础上，建立联系
  > 创建关系时，必须指定关系类型
  
  - 创建没有任何属性的关系
  
  ```
  match (a:Director), (b:Movie)
  where a.name = "张艺谋" AND b.title = "英雄"
  create (a)-[r:DIRECT]->(b)
  return r;
  
  ```
  
  用属性是没有办法唯一性标识实体的，如果有两个Robert，怎么准确匹配到唯一一个呢？
  
  - 创建关系，并设置关系属性
  
  ```
  match (a:Actor), (b:Movie)
  where a.name = "梁朝伟" AND b.title = "英雄"
  create (a)-[r:ACTED_IN{roles:"残剑"}]->(b)
  return r;
  
  match (a:Actor), (b:Movie)
  where a.name = "张曼玉" AND b.title = "英雄"
  create (a)-[r:ACTED_IN{roles:"飞雪"}]->(b)
  return r;
  
  match (a:Actor), (b:Actor)
  where a.name = "张曼玉" AND b.name = "梁朝伟"
  create (a)-[r:CP]-(b)
  return r;

  ```
  
4. 查询关系

  > 在Cypher中，关系分为三种：
    - --，表示有关系
    - --> / <--，表示有方向的关系

  - 查询整个数据库 match(n) return n;
  - 查询跟指定节点有关系的节点 match(n)--(m:Movie) return n;
  - 查询有向关系的节点 match(n) -->(n:Movie) return n;
  - 为关系命名，通过[r]为关系定义一个变量名，通过函数type获取关系的类型  match(n)-[r]->(:Movie) return r, type(r);
  - 查询特定的关系类型，通过[Variable:RelationshipType{Key:Value}]指定关系的类型和属性 match(n)-[r:ACTED_IN{roles:"飞雪"}]->(:Movie) return r, type(r);
 
5. 删除节点
  - 删除所有节点 match(n) delete n;

6. 删除关系
  - 删除关系 match (n)-[r]-() delete r;

7. 修改节点
  > 使用set，更新节点的标签和实体属性，remove移除实体的属性和节点的标签
  
  - 增加标签 match(n) where id(n)=0 set n:age return n;
  - 增加属性 match(n) where id(n)=0 set n.age = 62 return n;
  - 移除属性 match(n) where id(n)=0 remove n.age return n;
  - 移除标签 match(n) where id(n)=0 remove n:age return n;
  
8. 修改关系
  > 和节点类似，也可以为关系增加和删除属性
  
  - 增加属性 match (n)-[r]->(m) where id(n)=0 and id(m)=57 set r.duty="总导演" return n, r, m;
