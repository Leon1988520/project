## 四. Neo4j cookbook

### 4.1 Neo4j 安装与配置

> 以windows为例，如果不详尽之处，请参阅参考文献3

1. 下载安装JDK，并配置JDK路径，在dos下，键入java -version，显示java版本即表示JDK安装且配置成功
    > 笔者的JDK版本如下

    ```
    java version "1.7.0_67"
    Java(TM) SE Runtime Environment (build 1.7.0_67-b01)
    Java HotSpot(TM) 64-Bit Server VM (build 24.65-b04, mixed mode)
    ```

2. 下载Neo4j zip文件（个人使用，请下载community版本）；将其解压到目标文件夹下；在环境变量中配置Neo4j_HOME
3. 启动Neo4j程序
> 以管理员身份打开控制台(WIN+R, 键入cmd)，切到Neo4j的bin文件下，运行： neo4j.bat console

4. 把Neo4j安装为服务
    ```
    安装和卸载服务：
    bin\neo4j install-service
    bin\neo4j uninstall-service

    启动服务，停止服务，重启服务和查询服务的状态：
    bin\neo4j start
    bin\neo4j stop
    bin\neo4j restart
    bin\neo4j status

    ```
5. 进入Neo4j数据库界面
> 打开浏览器，键入Neo4j默认访问地址 http://localhost:7474/， 默认账号和密码均为neo4j
